# Model Building 

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import svm
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris=pd.read_csv("Iris.csv")
```

```python
iris.describe
```

# Spliting  The dataset 
(training data-80% and Testing data-20%)

```python
var_train,var_test=train_test_split(iris,test_size=0.20)
```

**Trainig data** 

```python
var_train
```

**Testing data**

```python
var_test
```

```python
train_X= var_train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]#attributes
```

```python
train_Y=var_train.Species#target
```

```python
train_X
```

**Target variable** 

```python
train_Y
```

```python
test_X=var_test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]#attributes
test_Y=var_test.Species#target
```

```python
#svm 
model=svm.SVC()
```

```python
model.fit(train_X,train_Y)
predict=model.predict(test_X)
```

```python
print("Accuracy :",sklearn.metrics.accuracy_score(predict,test_Y))
```

```python
#Decision Tree
model= DecisionTreeClassifier()
```

```python
model.fit(train_X,train_Y)
```

```python

predict=model.predict(test_X)
```

```python
print("Accuracy :",sklearn.metrics.accuracy_score(predict,test_Y))
```

**Conclusion:-- Two Model gives accuracy around 90%-95%**
